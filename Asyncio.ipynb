```python
limit = asyncio.Semaphore(3)

async def get_coordinates(session,coordinates):
    async with await session.post(url="https://geocode.xyz",data={"locate": coordinates,"geoit": "json"}) as r:
        address= r.json()
        if r.status == 200:
            return await address
        else:
            return None
async def get_all(session, coordinates):
    tasks = []
    for i, coordinate in enumerate(coordinates):
        task = asyncio.create_task(get_coordinates(session,coordinate))
        tasks.append(task)
    results = await asyncio.gather(*tasks)
    return results

async def main():
    async with aiohttp.ClientSession() as session:
        result = await get_all(session, coordinates)
        return result
```
